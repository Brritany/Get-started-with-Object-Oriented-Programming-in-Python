# 二、 常用模組呼叫方式

## 範例模組 calculate.py

```python
# calculate.py
def add(n1,n2):
    return n1 + n2
    
def sub(n1,n2):
    return n1 - n2
def mul(n1,n2):
    return n1 * n2
def div(n1,n2):
    return n1 / n2
```

## 呼叫模組(module)`calculate.py`中自訂函式方法

```python
# 匯入模組
import calculate

# 調用時須使用 模組名.自訂函式名
print(calculate.add(5,2))  # 7
print(calculate.sub(5,2))  # 3
```

```python
# 匯入模組內自訂函式，沒寫上的自訂函式將無法被調用
from calculate import add,sub

print(add(5,2))  # 7
print(sub(5,2))  # 3
```

```python
# 匯入模組內自訂函式，使用【*】可匯入全部函式
# 注意若函式名與其他模組中函式名重複，將會直接取代其他模組中的函式
from calculate import *

print(add(5,2))  # 7
print(sub(5,2))  # 3
```

```python
# 當自訂函式名稱太長時，可使用【as + 自訂別名】
# 後續調用時輸入自訂別名即可
import calculate as cal

print(cal.add(5,2))  # 7
print(cal.sub(5,2))  # 3
```
